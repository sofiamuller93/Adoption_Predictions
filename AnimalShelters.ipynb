```python
import pandas as pd
from datetime import datetime
import numpy as np
```

```python
intakes = pd.read_csv("Resources/Austin_Animal_Center_Intakes.csv")
outcomes = pd.read_csv("Resources/Austin_Animal_Center_Outcomes.csv")
```

```python
intakes.head()
```

```python
outcomes.head()
```

```python
outcomes_df = pd.DataFrame(outcomes)
intakes_df = pd.DataFrame(intakes)
```

```python
outcomes_df["Animal ID"].count()
```

```python
intakes_df["Animal ID"].count()
```

```python
animals = pd.merge(intakes_df, outcomes_df, how='left', on="Animal ID")
```

```python
animals.head()
```

```python
animals["Animal ID"].count()
```

```python
sex_outcome = animals["Sex upon Outcome"].str.split(' ', 1, expand=True)
fixed = sex_outcome[0]
gender = sex_outcome[1]
```

```python
type(fixed)
```

```python
animals["Gender"]=gender
animals["Fixed"]=fixed
```

```python
age_intake = animals["Age upon Intake"].str.split(' ', 1, expand=True)
number = age_intake[0]
multiple = age_intake[1]
```

```python
animals['Age Number'] = number
animals['Age Multiple'] = multiple
```

```python
dates = {'years': 365, 'year': 365, 'months': 30, 'month': 30, 'weeks': 7, 'week': 7, 'days': 1, 'day': 1, 'nan': 0}
```

```python
animals['Age Multiple'] = [dates[item] for item in animals['Age Multiple']]
```

```python
animals.head()
```

```python
age_outcome = animals["Age upon Outcome"].str.split(' ', 1, expand=True)
number_out = age_outcome[0]
multiple_out = age_outcome[1]

sex_intake = animals["Sex upon Intake"].str.split(' ', 1, expand=True)
fixed_intake = sex_intake[0]
gender_intake = sex_intake[1]
```

```python
animals['Age Number Outcome'] = number_out
animals['Age Multiple Outcome'] = multiple_out

animals['Fixed on Intake'] = fixed_intake
```

```python
animals.head()
```

```python
animals["Age Multiple Outcome"].fillna('zero', inplace=True)
dates = {'years': 365, 'year': 365, 'months': 30, 'month': 30, 'weeks': 7, 'week': 7, 'days': 1, 'day': 1, 'zero': 0}
```

```python
animals['Age Multiple Outcome'] = [dates[item] for item in animals['Age Multiple Outcome']]
animals.dtypes
```

```python
# pd.to_numeric(animals['Age Multiple'], errors='coerce')
# pd.to_numeric(animals['Age Number Outcome'], errors='coerce')
# pd.to_numeric(animals['Age Multiple Outcome'], errors='coerce')
# pd.to_numeric(animals['Age Number'], errors='coerce')
animals['Intake Number'] = animals['Age Number'].fillna(0).astype(int)
animals['Intake Multiple'] = animals['Age Multiple'].fillna(0).astype(int)

animals['Outcome Number'] = animals['Age Multiple Outcome'].fillna(0).astype(int)
animals['Outcome Multiple'] = animals['Age Number Outcome'].fillna(0).astype(int)
```

```python
# animals.dtypes
```

```python
animals.head()
```

```python
animals["Intake Date"] = pd.to_datetime(animals["DateTime_x"])
animals['Outcome Date'] = pd.to_datetime(animals["DateTime_y"])
```

```python
animals['Stay Duration'] = animals['Outcome Date'] - animals["Intake Date"] 
```

```python
animals.head()
```

```python
animals.dtypes
```

```python
animals.columns
```

```python
animals_final = pd.DataFrame({'Animal ID': animals['Animal ID'],
                             'Name': animals['Name_x'],
                             'Animal Type': animals['Animal Type_x'],
                             'Breed': animals['Breed_x'],
                             'Gender': animals['Gender'],
                             'Color': animals['Color_x'],
                             'Birthday': animals['Date of Birth'],
                             'Intake DateTime': animals['DateTime_x'],
                             'Intake Type': animals['Intake Type'],
                             'Intake Condition': animals['Intake Condition'],
                             'Intake Age (days)': animals['Intake Number']*animals['Intake Multiple'],
                             'Fixed upon Intake': animals['Fixed on Intake'],
                             'Outcome Type': animals['Outcome Type'],
                             'Outcome Subtype': animals['Outcome Subtype'],
                             'Outcome Date': animals['DateTime_y'],
                             'Fixed upon Outcome': animals['Fixed'],
                             'Outcome Age': animals['Outcome Number']*animals['Outcome Multiple'],
                             'Stay Duration': animals['Stay Duration']
        
})
animals_final.head()
```

```python
animals_final.to_csv('Animals.csv')
```
